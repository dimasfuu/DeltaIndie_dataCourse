```python
# 1. Hitung rata rata penjualan dari produk?
df.loc[df["Jenis Produk"]=='Ivory230',["Jumlah Order"]].mean()
```

```python
# 2. Di tanggal berapakah transaki termahal terjadi?
df.loc[df["Jenis Produk"]=='Ivory230',["Tanggal","Total"]].max()
```

```python
# 3. Berapakah jumlah orderan terbanyak di produk kamu?
df.loc[df["Jenis Produk"]=='Ivory230',["Tanggal","Jumlah Order"]].max()
```

```python
# 4. Berapa nilai terendah total pembelian yang terjadi di bulan 10 tahun 2023?
df.loc[(df["Tanggal"] >= "2023/10/01") & (df["Tanggal"] <= "2023/10/30"),["Jumlah Order"]].min()
```

```python
# 5. Di tanggal berapa nilai harga barang mu tertinggi?
df.loc[df["Jenis Produk"]=='Ivory230',["Tanggal","Harga"]].max()
```
