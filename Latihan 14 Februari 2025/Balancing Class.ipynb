# 4. Balancing

## 4.1 Balancing(OVERSAMPLING)

```python
#masih data asli (bukan data latih)
x = df_new2.drop(columns=["Outcome"])
y = df_new2["Outcome"]
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from collections import Counter
#memanggil SMOTE
smote = SMOTE(random_state=42)
```

```python
#kode oversampling

#Split data menjadi Train (80%) dan Test (20%)*
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

# *Oversampling dengan SMOTE (hanya pada training data)*
smote = SMOTE(random_state=42)
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
print("Distribusi kelas setelah SMOTE (Oversampling):", Counter(y_train_smote))
```

```python
#kode oversampling

#Split data menjadi Train (70%) dan Test (30%)*
x_train70, x_test30, y_train70, y_test30 = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)

# *Oversampling dengan SMOTE (hanya pada training data)*
smote = SMOTE(random_state=42)
x_train_smote7030, y_train_smote7030 = smote.fit_resample(x_train70, y_train70)
print("Distribusi kelas setelah SMOTE (Oversampling):", Counter(y_train_smote7030))
```

## 4.2 Balancing(UNDERSAMPLING)

```python
#masih data asli (bukan data latih)
x = df_new2.drop(columns=["Outcome"])
y = df_new2["Outcome"]
```

```python
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import train_test_split
from collections import Counter
#memanggil SMOTE
smote = SMOTE(random_state=42)
```

```python
#kode oversampling

#Split data menjadi Train (80%) dan Test (20%)*
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

# *Oversampling dengan SMOTE (hanya pada training data)*
rus = RandomUnderSampler(random_state=42)
x_train_rus, y_train_rus = rus.fit_resample(x_train, y_train)
print("Distribusi kelas setelah Random Under Sample (Underrsampling):", Counter(y_train_rus))
```

```python
#kode oversampling

#Split data menjadi Train (70%) dan Test (30%)*
x_train70, x_test30, y_train70, y_test30 = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)

# *Oversampling dengan SMOTE (hanya pada training data)*
rus = RandomUnderSampler(random_state=42)
x_train_rus7030, y_train_rus7030 = rus.fit_resample(x_train70, y_train70)
print("Distribusi kelas setelah Random Under Sample (Underrsampling):", Counter(y_train_rus7030))
```

# 5. Training Testing ML

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

### 5.1 Buat Training KNN

```python
# memuat model KNN 
model_knn = KNeighborsClassifier(n_neighbors=5)

#melatih model machine learning
model_knn.fit(x_train_smote,y_train_smote)

#menampilkan uji performa sisi training
y_pred_train_KNN = model_knn.predict(x_train_smote)

# menampilkan uji performa sisi testing
y_pred_test_KNN = model_knn.predict(x_test)
```

```python
y_test
```

## 5.2 Metriks Evaluasi Model KNN

```python
# menampilkan akurasi sisi training
akurasi_traning_knn = accuracy_score(y_train_smote,y_pred_train_KNN)

# menampilkan akurasi sisi testing
akurasi_test_knn = accuracy_score(y_test,y_pred_test_KNN)

print("Hasil akurasi trainingnya adalah : ", akurasi_traning_knn*100)
print("Hasil akurasi testingnya adalah : ",akurasi_test_knn*100)
```

```python
print("Classification report training:", classification_report(y_train_smote,y_pred_train_KNN))
```

```python
print("Classification report testing:", classification_report(y_test,y_pred_test_KNN))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Confusion Matrix-training
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_train_smote, y_pred_train_KNN), annot=True, fmt='d', cmap='Blues', xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

# 6. Memprediksi model baru (Program prediksi)

```python
def prediksi_diabetes(model) : 
    print("=== Silakan masukkan data pasien ===")
    Pregnancies = int(input("Masukan Pregnancies : "))
    Glucose = int(input("Masukan Kadar Glukosa : "))
    BloodPressure = int(input("Masukan Tekanan Darah : "))
    SkinThickness = int(input("Masukan Ketebalan Kulit : "))
    Insulin = int(input("Masukan Kadar Insulin : "))
    BMI = float(input("Masukan BMI Pasien : "))
    DiabetesPedigreeFunction = float(input("Masukan DPF Pasien"))
    Age = int(input("Masukan umur Pasien : "))
    
    data_baru = np.array([Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age])
    data_baru = data_baru.reshape(1,-1)
    prediksi_baru = model.predict(data_baru)
    
    if prediksi_baru ==1:
        print("Pasien terkena Diabetes")
    else :
        print("Pasien tidak terkena Diabetes")
 
```
