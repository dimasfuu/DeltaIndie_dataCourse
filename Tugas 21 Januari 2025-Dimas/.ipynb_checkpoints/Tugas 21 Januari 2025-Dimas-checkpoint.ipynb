## 2. Data Cleaning

### 2.1. Transformasi Variabe/Kolom

Sebelum melakukan proses cleaning missing values diperlukan beberapa perbaikan pada kolom yang diperlukan, seperti **transformasi tipe data kolom 'Tanggal'**, dan **penambahan Kolom 'Tahun' dan 'Bulan'** agar data dapat dieksplorasi lebih lanjut

#### a. Mengubah tipe data Kolom 'Tanggal'

```python
data_jualan["Tanggal"] = pd.to_datetime(data_jualan["Tanggal"], format = "%d/%m/%Y")
data_jualan
```

```python
data_jualan.info()
```

**Kolom Tanggal sudah terbaca sebagai tipe data "Date time" dan dapat kita lanjutkan dengan menambahkan kolom 'Bulan' dan 'Tahun'.**

#### b. Menambahkan kolom "Bulan" dan "Tahun" dari kolom "Tanggal"

```python
data_jualan["Bulan"] = data_jualan["Tanggal"].dt.month_name()
data_jualan["Tahun"] = data_jualan["Tanggal"].dt.year
data_jualan
```

agar lebih memudahkan dalam proses pemanggilan dataset, untuk selanjutnya dataset **"data_jualan"** akan disingkat dengan **"dj"**.

### 2.2. Data cleaning for missing values

Dari dataset **"data_jualan"** / **"dj"** prosess data cleaning untuk data hilang/"missing values" dapat dilakukan dengan metode replacing, adapun perlakuan metode replacing yang dilakukan antara lain : 

1. Mengganti/mengisi data hilang untuk kolom 'Jenis Produk' dengan metode pengambilan data 'Modus'
2. Mengganti/mengisi data hilang untuk kolom 'Harga' dengan mengisi data harga sesuai dengan hasil 'Harga' satuan pada Jenis Produk yang bersesuaian.
3. Mengganti/mengisi data hilang untuk kolom 'Jumlah order' dengan metode pengambilan data 'Mean/Rata-rata'
